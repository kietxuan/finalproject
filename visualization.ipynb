```python
import pandas as pd
import matplotlib.pyplot as plt
# Read the CSV file
data = pd.read_excel(r"C:\Users\MY PC\Documents\project data\age_of_customer.xlsx")
print(data.head())
plt.bar(data['age_group'], data['number of customers'], color='skyblue')

# Thêm tiêu đề và nhãn
plt.title('Number of customers by age')
plt.xlabel('Age')
plt.ylabel('Number of customers')

# Xoay nhãn trục X nếu cần cho dễ đọc
plt.xticks(rotation=45)

# Hiển thị biểu đồ
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc file
data = pd.read_excel(r"C:\Users\MY PC\Documents\project data\111.xlsx")

# Chuẩn bị dữ liệu
product_sales = data.groupby('product_type_name')['purchase_count'].sum().sort_values(ascending=False)
color_sales = data.groupby('perceived_colour_master_name')['purchase_count'].sum().sort_values(ascending=False)

# Tạo subplots
fig, axes = plt.subplots(2, 1, figsize=(10, 10))  # 2 hàng, 1 cột

# Vẽ biểu đồ cột - loại sản phẩm
product_sales.plot(kind='bar', color='skyblue', ax=axes[0])
axes[0].set_title('Top 5 products purchased')
axes[0].set_xlabel('Product type')
axes[0].set_ylabel('Total number of purchases')
axes[0].tick_params(axis='x', rotation=45)

# Vẽ biểu đồ tròn - màu sắc
color_sales.plot(kind='pie', autopct='%1.1f%%', ax=axes[1])
axes[1].set_ylabel('')
axes[1].set_title('Top colors loved by users')

plt.tight_layout()
plt.show()
```
